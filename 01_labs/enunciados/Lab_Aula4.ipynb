## 1. *Series*



<p align="justify">  Em <i>Pandas</i>, uma <i>Series</i> é uma estrutura de dados unidimensional que se assemelha a uma coluna em uma planilha ou a um vetor em programação. Cada elemento em uma <i>Series</i> possui um rótulo associado, chamado de índice. Essa combinação de dados e índices torna as <i>Series</i> poderosas e flexíveis para trabalhar com dados em Python.

<p align="justify"> Aqui estão alguns pontos-chave sobre as <i>Series</i> em <i>Pandas</i>:


* <p align="justify">Unidimensional: As <i>Series</i> são estruturas de dados unidimensionais, o que significa que elas contêm apenas uma sequência de dados. Enquanto os <i>Data Frames</i>, outra estrutura de dados <i>Pandas</i>, são bidimensionais, as <i>Series</i> são projetadas para armazenar dados unidimensionais, como uma coluna de valores.

* <p align="justify">Índices: Cada elemento em uma <i>Series</i> tem um rótulo de índice exclusivo. Esses índices são automaticamente atribuídos às posições dos elementos na <i>Series</i> quando você a cria. Você também pode definir índices personalizados se desejar.

* <p align="justify">Tipos de Dados: As <i>Series</i> podem conter dados de diversos tipos, como números inteiros, números de ponto flutuante, strings, datas e assim por diante. O <i>Pandas</i> permite que as <i>Series</i> armazenem dados heterogêneos, ou seja, dados de diferentes tipos em uma única <i>Series</i>.

* <p align="justify">Operações de Vetorização: Uma das vantagens das <i>Series</i> é que elas suportam operações de vetorização. Isso significa que você pode aplicar operações diretamente a uma <i>Series</i> sem precisar de loops explícitos. Isso torna o código mais eficiente e legível.

* <p align="justify">Acesso aos Dados: Você pode acessar os elementos de uma <i>Series</i> de várias maneiras, incluindo índices numéricos, índices de rótulos, fatiamento e seleção condicional. Isso facilita a recuperação de dados específicos ou a realização de operações neles.

* <p align="justify">Nome da <i>Serie</i>: Você pode atribuir um nome a uma <i>Series</i>, o que pode ser útil para documentação e referência. O nome é armazenado como um atributo da <i>Series</i>.



<p align="justify"> Para criar uma <i>Series</i> em <i>Pandas</i>, pode-se fazer de forma similar com os <i>NumPy</i> arrays:

````python
import pandas as pd

# Criando uma série a partir de uma lista

dados = [1,2,3,4,5]

serie = pd.Series(dados)

print(serie)
print(type(serie))
````

<p align="justify"> Veja que o valor impresso é diferente de uma lista ou de um <i>NumPy</i> array: automaticamente são criados índices para cada elemento da <i>Serie</i>. As <i>Series</i> são uma parte fundamental do <i>Pandas</i> e são frequentemente usadas em conjunto com <i>Data Frames</i> para realizar análises e manipulações de dados mais detalhadas. Elas são versáteis e desempenham um papel crucial em muitas operações de análise de dados em Python. Se quiser definir o índice a ser utilizado (ao invés de um índice que começa do 0):


````python
dados2 = pd.Series([10,7,3,2,-15], # Lista de valores
                   index = ['Brasil', 'Japão', 'Coréia', 'México', 'EUA']) # Valores do índice

print(dados2)
````

<p align="justify"> De forma parecida com arrays do <i>NumPy</i>, você pode usar os labels para selecionar valores:

````python
dados2["Brasil"]
````

Ou uma lista de valores:

````python
dados2[['Brasil', 'México', 'EUA']]
````

Você pode filtrar valores utilizando operações Booleanas:

````python
dados2[dados2 > 5]
````

Pode fazer operações matemáticas:

````python
dados2 * 2
````

<p align="justify"> E assim por diante. Além disso, como ressalta Mckinney (2022), <i>Series</i> de <i>Pandas</i> são parecidas também com dicionários. Você pode verificar se determinada "chave" (ou índice nesse contexto) está dentro de uma <i>Serie</i>:

````python
"Brasil" in dados2
````

````python
"Argentina" in dados2
````

<p align="justify">Se passar um dicionário para transformar em <i>Serie</i>, automaticamente o <i>Pandas</i> utilizará as chaves como índice:

````python
dados_dict = {'Brasil': 10, 'Japão': 7, 'Coréia': 3, 'México':2, 'EUA': -15}

dados3 = pd.Series(dados_dict)

dados3
````

<p align="justify"> Quando existem valores <i>NA</i> (ou <i>missing</i>), você pode usar o método isna() para identificá-los:

````python
import pandas as pd
import numpy as np

# Criando uma Series com índice de países
dados4 = {'Brasil': 210_000_000,
          'EUA': 328_000_000,
          'China': 1_409_000_000}

series2 = pd.Series(dados4)

# Definindo um valor ausente para um país
series2['Índia'] = np.nan

# Imprimindo a série
print(series2)

print(" ")
print("*****Missings******")


# Checando os ausentes
print(series2.isna())
````

Por fim, se quiser pode nomear tanto os índices da série quanto seus valores:

````python
# Nomeando os valores

series2.name = 'PIB per capita'

# nomeando o índice

series2.index.name = 'Países'


print(series2)
````

## 2. <i>Data Frames</i>

<p align="justify">Um <i>Data Frame</i> é uma estrutura de dados bidimensional, tabular e flexível amplamente usada na biblioteca <i>Pandas</i> do Python para armazenar e manipular dados de forma semelhante a uma planilha ou tabela em um banco de dados. Ele é composto por linhas e colunas, onde cada coluna pode conter dados de diferentes tipos, como números inteiros, números de ponto flutuante, strings, datas, entre outros.

<p align="justify">Aqui estão algumas características e conceitos importantes relacionados a <i>Data Frames</i>:

* <p align="justify">Bidimensional: <i>Data Frames</i> são bidimensionais, o que significa que eles têm duas dimensões: linhas e colunas. Cada linha representa um registro ou entrada de dados, enquanto cada coluna representa uma variável ou atributo específico.

* <p align="justify">Estrutura Tabular: Os dados em um <i>Data Frame</i> são organizados em uma estrutura tabular semelhante a uma planilha. As colunas são nomeadas e as linhas são numeradas, permitindo fácil indexação e identificação dos dados.

* <p align="justify">Tipos de Dados Heterogêneos: Diferentemente das estruturas de arrays <i>NumPy</i>, em que todos os elementos devem ser do mesmo tipo de dados, <i>Data Frames</i> do <i>Pandas</i> permitem que cada coluna contenha dados de diferentes tipos. Isso é particularmente útil para lidar com conjuntos de dados do mundo real, onde os tipos de dados variam.

* <p align="justify">Rótulos de Linha e Coluna: Os <i>Data Frames</i> têm rótulos de linha (índices) e rótulos de coluna. Os rótulos de linha são usados para identificar as entradas de dados, enquanto os rótulos de coluna são usados para nomear as variáveis ou atributos.

* <p align="justify">Operações de Dados: Os <i>Data Frames</i> oferecem uma ampla gama de funcionalidades para realizar operações de manipulação de dados, como filtragem, seleção, ordenação, agregação e cálculos estatísticos.

* <p align="justify">Integração com <i>Pandas</i>: Os <i>Data Frames</i> são a estrutura de dados central na biblioteca <i>Pandas</i>, o que os torna altamente compatíveis com outras funcionalidades do <i>Pandas</i>, como leitura e escrita de dados em diferentes formatos, manipulação de datas e horas, e visualização de dados.

* <p align="justify">Origem de Dados Diversificada: <i>Data Frames</i> podem ser criados a partir de várias fontes de dados, como arquivos CSV, bancos de dados SQL, dicionários Python e muito mais. Eles também podem ser exportados para diversos formatos, tornando-os ideais para análise e ciência de dados.

<p align="justify"> Em resumo, um <i>Data Frame</i> é uma estrutura de dados extremamente versátil e poderosa no <i>Pandas</i>, projetada para lidar com uma ampla variedade de tarefas de manipulação e análise de dados. É uma escolha popular para qualquer pessoa que trabalhe com dados em Python devido à sua flexibilidade e eficiência.

<p align="justify">Para criar um <i>data frame</i>, você pode utilizar um dicionário com listas aninhadas:

````python
# Criando uma lista de dicionários com dados fictícios
dados = [
    {'Nome': 'Alice', 'Idade': 25, 'Cidade': 'São Paulo'},
    {'Nome': 'João', 'Idade': 30, 'Cidade': 'Rio de Janeiro'},
    {'Nome': 'Carol', 'Idade': 22, 'Cidade': 'Belo Horizonte'},
    {'Nome': 'David', 'Idade': 35, 'Cidade': 'São Paulo'},
]

# Criando um Data Frame a partir da lista de dicionários
df = pd.DataFrame(dados)

# Exibindo o DataFrame
print(df)
````

<p align="justify"> Nem sempre você vai querer imprimir todo os dados. Digamos que você queira só entender mais ou menos o formato do banco. No exemplo a seguir temos um <i>data frame</i> com 20 observações:

````python
import random

# Criando dados fictícios para o Data Frame
dados = {
    "Nome": ["Pessoa" + str(i) for i in range(1, 21)], # usando list comprehension para gerar um índice de 1 a 21 e somar à "Pessoa"
    "Idade": [random.randint(18, 65) for _ in range(20)], # usando list comprehension para gerar números aleatórios no intervalo de 18 a 65 para 20 observações
    "Cidade": ["Cidade" + str(random.randint(1, 5)) for _ in range(20)]  # usando list comprehension para gerar um índice de 1 a 5 e somar à "Cidade"
}

# Criando o Data Frame
df = pd.DataFrame(dados)

# Exibindo o Data Frame
print(df)
````

##### Loc e Iloc

<p align="justify"><i>loc</i> e <i>iloc</i> são dois atributos fundamentais para a indexação e seleção de dados em <i>Pandas Data Frames</i> e <i>Series</i>. Eles são usados para acessar elementos específicos com base em rótulos (para <i>loc</i>) ou índices inteiros (para <i>iloc</i>). Vamos explorar cada um deles em detalhes.

###### <i>loc</i> (localização):

* <p align="justify"><i>loc</i> é usado para acessar dados com base em rótulos ou rótulos de índices, em vez de posições numéricas.
* <p align="justify">Ele permite que você selecione dados com base em nomes de colunas ou rótulos de índices, usando uma sintaxe semelhante à da notação de dicionário.
* <p align="justify">O primeiro argumento de <i>loc</i> é o rótulo das linhas que você deseja selecionar, e o segundo argumento é o rótulo das colunas (opcional).
* <p align="justify">Pode ser usado para seleções condicionais e fatiamento baseado em rótulos.

````python
# Exemplo da documentação do loc

df = pd.DataFrame({
    "Animal": ['cobra', 'viper', 'sidewinder'],
    "Velocidade": [1,4,7],
    "Escudo": [2,5,8]
})

df
````

Para pegar um label só:

````python
# Pegando a primeira observação

df.loc[0]
````

Se fornecer uma lista de índices, o retorno da função será um <i>data frame</i>:

````python
df.loc[[0, 1]]
````

<p align="justify">Se o índice do <i>data frame</i> fosse o nome dos animais, você poderia extrair diretamente utilizando o <i>label</i>:

````python
df = pd.DataFrame([[1, 2], [4, 5], [7, 8]],
     index=['cobra', 'viper', 'sidewinder'],
     columns=['max_speed', 'shield'])

df
````

````python
df.loc['cobra']
````

````python
df.loc[['cobra','viper']]
````

E assim por diante...

###### <i>Iloc</i> (indexação da localização)

* <p align="justify"><i>iloc</i> é usado para acessar dados com base em posições numéricas de linhas e colunas, semelhante ao uso de índices em listas.
* <p align="justify">Ele não considera rótulos, apenas índices inteiros.
* <p align="justify">O primeiro argumento de <i>iloc</i> é o índice da linha que você deseja selecionar e  o segundo argumento é o índice da coluna (opcional).
* <p align="justify">Pode ser usado para seleções condicionais e fatiamento baseado em índices inteiros.

````python
dados = {
    "Nome": ["Alice", "Bob", "Carol", "David", "Eva"],
    "Idade": [25, 30, 22, 35, 28],
    "Cidade": ["São Paulo", "Rio de Janeiro", "Belo Horizonte", "São Paulo", "Rio de Janeiro"]
}

df = pd.DataFrame(dados)

df
````

````python
# Selecionando a primeira linha
df.iloc[0]
````

````python

# Selecionando a segunda célula (0,1)
df.iloc[0, 1]

````

````python
# Selecionando as três primeiras linhas e todas as colunas
df.iloc[:3, :]
````

* 3 - Discretização e *Binning*:

<p align="justify">A discretização é o processo de dividir variáveis contínuas em intervalos ou categorias discretas. Você pode usar o método <i>cut()</i> para criar intervalos (bins) em seus dados. Suponha que temos um <i>Data Frame</i> com informações sobre idades de pessoas e queremos criar faixas etárias com base nas idades.



````python
dados = {
    'Nome': ['Alice', 'Bob', 'Carol', 'David', 'Eva'],
    'Idade': [25, 30, 22, 35, 28]
}

df = pd.DataFrame(dados)

# Definindo os limites dos bins e as categorias correspondentes
limites_bins = [0, 18, 25, 35, float('inf')]  # Faixas etárias: 0-18, 19-25, 26-35, 36+
categorias = ['0-18', '19-25', '26-35', '36+']

# Usando cut() para criar uma nova coluna 'Faixa Etária' com base nas idades
df['Faixa Etária'] = pd.cut(df['Idade'], bins=limites_bins, labels=categorias)

print(df)
````

<p align="justify">Neste exemplo, usamos o método <i>cut()</i> para criar uma nova coluna chamada 'Faixa Etária' com base nas idades das pessoas. Definimos os limites dos bins (faixas etárias) em 'limites_bins' e as categorias correspondentes em 'categorias'. O método <i>cut()</i> atribui automaticamente a cada idade a categoria apropriada com base nos limites especificados.

<p align="justify">Isso é útil quando você deseja transformar variáveis contínuas, como idades, em categorias discretas para análise. O <i>Pandas</i> faz o trabalho pesado de atribuir os valores corretos a cada categoria com base nos limites especificados.

* 4 - Criando variáveis *dummy* (ou binárias)

<p align="justify">Elas são usada quando você deseja converter variáveis categóricas em um formato numérico que possa ser utilizado em algoritmos de machine learning ou em análises estatísticas que requerem entradas numéricas.

<p align="justify">As variáveis dummy são variáveis binárias (0 ou 1) que representam a presença ou ausência de uma característica específica em uma variável categórica. Variáveis dummy são usadas quando você tem variáveis categóricas (como cores, países, categorias de produtos, etc.) que não podem ser diretamente usadas em algoritmos de machine learning ou análises estatísticas. Algoritmos e modelos matemáticos requerem entradas numéricas. Para calcular variáveis dummy, você cria uma nova coluna binária para cada categoria da variável categórica original. Cada coluna binária representa uma categoria específica e é preenchida com 1 (verdadeiro) se a categoria estiver presente e 0 (falso) caso contrário.

<p align="justify">A biblioteca <i>Pandas</i> do Python possui funções para criar variáveis dummy, como pd.get_dummies(). Exemplo:

````python
# Dados de exemplo
dados = {'Cor': ['Vermelho', 'Verde', 'Azul', 'Vermelho', 'Azul', 'Verde', 'Verde']}

# Criando um DataFrame a partir dos dados
df = pd.DataFrame(dados)

# Exibindo o DataFrame original
print("DataFrame Original:")
print(df)

# Calculando as variáveis dummy
df_dummies = pd.get_dummies(df, columns=['Cor'], prefix=['Cor'])

# Exibindo o DataFrame com as variáveis dummy
print("\nDataFrame com Variáveis Dummy:")
print(df_dummies)
````

#### Expressões Regulares

<p align="justify">Expressões regulares, também conhecidas como "regex" ou "regexp", são uma poderosa ferramenta para manipulação de strings em Python (e em muitas outras linguagens de programação). As expressões regulares permitem realizar pesquisa, correspondência e manipulação de padrões de texto complexos de maneira eficiente. O módulo <i>re</i> do Python é amplamente utilizado para trabalhar com expressões regulares. Aqui estão alguns conceitos importantes relacionados a expressões regulares em Python:


* 1 - Padrões de Correspondência:

<p align="justify">As expressões regulares são usadas para especificar padrões de texto que você deseja encontrar em uma string. Por exemplo, você pode usar uma expressão regular para encontrar todas as ocorrências de endereços de e-mail em um texto.

* 2 - Módulo <i>re</i>:

<p align="justify">O módulo <i>re</i> é a biblioteca padrão do Python para trabalhar com expressões regulares. Ele fornece funções e classes para compilar e usar expressões regulares.

* 3 - Funções Principais do Módulo <i>re</i>:

re.match(): Verifica se o padrão ocorre no início da string.

re.search(): Procura por qualquer local onde o padrão ocorre na string.

re.findall(): Retorna todas as ocorrências do padrão na string.

re.finditer(): Retorna um iterador de todas as correspondências do padrão na string.

re.sub(): Substitui todas as ocorrências do padrão por outra string.

* 4 - Padrões Básicos:

<p align="justify">Os padrões de expressões regulares podem incluir caracteres literais (como letras e números) e caracteres especiais que têm significados especiais (por exemplo, . para qualquer caractere, * para zero ou mais ocorrências, [ ] para especificar um conjunto de caracteres, etc.).

* 5 - Caracteres de Escape:

<p align="justify">Se você precisar corresponder a caracteres especiais como ., *, [, ], etc., como literais, é necessário "escapá-los" usando a barra invertida \.

* 6 - Exemplos de Expressões Regulares:

<p align="justify">Alguns exemplos de padrões de expressões regulares comuns incluem:

  \d+: Corresponde a um ou mais dígitos.

  [A-Za-z]+: Corresponde a uma ou mais letras maiúsculas ou minúsculas.

  .*: Corresponde a qualquer sequência de caracteres.

  \b\w+\b: Corresponde a palavras completas em uma string.

  [0-9]{2,4}: Corresponde a um número que tem de 2 a 4 dígitos.

* 7 - Exemplo de Uso:

<p align="justify">Aqui está um exemplo simples de uso do módulo <i>re</i> para encontrar endereços de e-mail em uma string:

````python
import re

texto = "Meu email é joao@exemplo.com e o outro é maria@exemplo.org"
padrao = r'\S+@\S+'

emails = re.findall(padrao, texto)
print(emails)
````

* 8 - Flags:

<p align="justify">O módulo <i>re</i> permite usar flags para modificar o comportamento das expressões regulares. Por exemplo, a flag re.IGNORECASE torna a correspondência de texto maiúsculo/minúsculo insensível.

<p align="justify">Expressões regulares podem ser incrivelmente úteis para análise de texto, validação de entradas, extração de informações e muito mais. No entanto, elas também podem ser complexas e difíceis de ler. Portanto, é aconselhável estudar e praticar expressões regulares para aproveitar ao máximo essa poderosa ferramenta.


####  Funções de Strings em Pandas

<p align="justify">O <i>Pandas</i> oferece uma variedade de funções para manipulação de strings em <i>Data Frames</i>. Essas funções são extremamente úteis quando você precisa trabalhar com dados de texto em colunas de um <i>Data Frame</i>. Aqui estão algumas das funções de string mais comuns em <i>Pandas</i>, juntamente com exemplos:

* 1-  str.lower() e str.upper():

<p align="justify">Converte todos os caracteres para minúsculas ou maiúsculas, respectivamente.
Exemplo:

````python
import pandas as pd

data = {'Nome': ['Alice', 'Bob', 'Carol', 'David'],
        'Cidade': ['New York', 'Los Angeles', 'San Francisco', 'Chicago']}
df = pd.DataFrame(data)

df['Cidade'] = df['Cidade'].str.lower()
print(df)
````

* 2 - str.len():

Verifica o tamanho, em caracteres, de uma string:

````python
df['Tamanho do Nome'] = df['Nome'].str.len()
print(df)
````

* 3 - str.contains():

<p align="justify">Verifica se uma string contém uma determinada substring e retorna uma série de booleanos:

````python
df['Contém "San"'] = df['Cidade'].str.contains('san', case=False)
print(df)
````

* 4 - str.replace():

Substitui uma substring por outra em cada string da coluna:

````python
df['Cidade'] = df['Cidade'].str.replace(' ', '-') # Substituindo espaço vazio por '-'
print(df)
````

* 5 - str.extract():

Extrai informações usando expressões regulares:

````python
df['Estado'] = df['Cidade'].str.extract(r'([A-Za-z]+)') # é um pouco obscuro ainda, mas futuramente retomaremos expressões regulares
print(df)
````
