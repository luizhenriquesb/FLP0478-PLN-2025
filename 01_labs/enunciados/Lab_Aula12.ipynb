Deep learning é uma subárea do Aprendizado de Máquina (machine learning) que utiliza redes neurais artificiais profundas para realizar tarefas complexas de aprendizado e tomada de decisão. O termo "profundo" refere-se ao uso de várias camadas (profundidade) em uma rede neural.

A ideia fundamental por trás do deep learning é inspirada no funcionamento do cérebro humano. Assim como o cérebro tem bilhões de neurônios interconectados que processam informações em camadas, as redes neurais artificiais profundas têm uma estrutura semelhante. Cada camada da rede consiste em neurônios (também chamados de unidades ou nodos) que realizam cálculos sobre os dados de entrada, transmitindo a informação através das camadas sucessivas até a camada de saída, onde a resposta final é gerada.

O treinamento de uma rede neural profunda envolve a apresentação de um conjunto de dados de entrada rotulado ao modelo, ajustando os pesos das conexões entre os neurônios para minimizar a diferença entre as previsões do modelo e as respostas reais. Esse processo é realizado através de algoritmos de otimização, como o gradiente descendente.

Deep learning tem se destacado em diversas aplicações, incluindo reconhecimento de padrões em imagens e vídeos, processamento da língua natural, reconhecimento de fala, jogos, medicina, entre outros. O aumento da capacidade computacional e o acesso a conjuntos de dados massivos têm contribuído para o sucesso e popularização do deep learning nos últimos anos. vamos explorar a arquitetura e o funcionamento das redes neurais, que são a base para modelos mais complexos.





### Redes Neurais

1. Neurônios e Suas Conexões:

Em uma rede neural, os neurônios são unidades fundamentais. Cada neurônio recebe inputs, realiza um cálculo ponderado, aplica uma função de ativação e produz um output. Esses neurônios são organizados em camadas, que são a base da estrutura de uma rede neural.

2. Arquitetura de uma Única Camada:

Uma única camada, também chamada de camada de entrada, consiste nos inputs da rede. Cada neurônio nesta camada representa uma feature do conjunto de dados. Essa camada é conectada diretamente à camada de saída, formando uma estrutura básica.

3. Funções de Ativação:

As funções de ativação, como a sigmoide ou a ReLU (Rectified Linear Unit), introduzem não-linearidades na rede, permitindo que ela aprenda padrões mais complexos nos dados. Essas funções são essenciais para o sucesso do treinamento, pois ajudam a modelar relações não lineares.

4. Redes Neurais Multicamadas (MLP):


Uma rede neural multicamadas, ou *multilayer perceptron* (MLP) consiste em uma camada de entrada, duas ou mais camadas ocultas e uma camada de saída. Cada conexão entre neurônios tem um peso associado, e o treinamento da rede envolve ajustar esses pesos para otimizar a performance do modelo.

O treinamento de uma MLP é realizado através de um processo chamado backpropagation. Durante o treinamento, a rede compara suas previsões com os valores reais, calcula a diferença (erro) e ajusta os pesos para reduzir esse erro. Este processo é repetido iterativamente até que a rede atinja um desempenho satisfatório.

Backpropagation é um algoritmo de otimização que utiliza o gradiente descendente para minimizar a função de custo. Ele propaga o erro da camada de saída de volta às camadas anteriores, ajustando os pesos em conformidade. Esse processo é crucial para que a rede aprenda padrões nos dados. ### Conceitos Principais



#### *Backpropagation*

Backpropagation é o principal algoritmo em redes neurais, que é amplamente utilizado para ajustar os pesos e vieses de uma rede neural de forma a minimizar o erro na saída. Em 1970, Seppo Linnainmaa introduziu uma técnica chamada de diferenciação automática reversa, capaz de calcular gradientes automaticamente e de forma eficiente. Esse processo, que posteriormente ficou conhecido como *backpropagation*, consiste em duas etapas principais: uma "passada para frente" (forward pass), na falta de uma tradução, em que as entradas são propagadas camada por camada até a saída; e uma "passada para trás" (backward pass), onde o erro é propagado em sentido inverso para calcular os gradientes de cada peso e viés. Esses gradientes indicam o quanto cada peso e viés deve ser ajustado para reduzir o erro da rede.

Assim, em cada iteração, o algoritmo calcula a contribuição de cada conexão para o erro e, com base nesses valores, realiza um passo de descida do gradiente, ajustando os parâmetros para minimizar o erro. Repetindo esse processo, a rede aprende a "sintonizar" os pesos de modo que o erro diminua progressivamente, até alcançar um mínimo.



#### Funções de Ativação

As funções de ativação desempenham um papel crucial em redes neurais artificiais. Sua função básica é definir se e como um neurônio será "ativado" ou não, dependendo do valor da entrada recebida. Em outras palavras, a função de ativação aplica uma transformação nos sinais de entrada, permitindo que a rede aprenda padrões complexos e representações mais sofisticadas. No entanto, não é exclusivo de métodos de aprendizado profundo. Uma função de ativação que já vimos foi a Sigmoide, que utilizamos em modelos de regressão logística.

Tipos comuns de funções de ativação:

* **ReLU** (Rectified Linear Unit): Retorna zero para entradas negativas e o próprio valor para entradas positivas. ReLU é amplamente utilizada devido à sua simplicidade e eficácia em redes profundas, ajudando a evitar problemas de saturação e facilitando o treinamento. Além disso, por não ter um valor limite para o *output*, ajuda a reduzir alguns problemas durante a "descida do gradiente" (*gradient descent*). Confira os capítulos 10 e 11 do Aurélion Géron.

* **Sigmoid**: Transforma a entrada em um valor entre 0 e 1, sendo útil em problemas onde queremos uma interpretação probabilística. Porém, sofre de saturação para valores extremos e pode dificultar o treinamento em redes mais profundas.

* **Tanh**: Transforma a entrada em um valor entre -1 e 1, ampliando a faixa de ativação em relação à Sigmoid, o que pode ajudar a minimizar a saturação. É uma escolha comum em redes com menos camadas.

#### Loss

Uma função de perda (*loss function*) é uma função que calcula um único valor numérico que o processo de aprendizado tentará minimizar. O cálculo da perda geralmente envolve calcular a diferença entre as saídas desejadas para alguns exemplos de treinamento e as saídas realmente produzidas pelo modelo quando esses exemplos são fornecidos (Stevens, Antiga e Viehmann, 2020, p. 109). Então, quanto menor a *loss*, melhor. E a *loss* de treinamento não deve ser muito menor do que a de validação, o que indicaria *overfitting*. A loss é otimizada por meio do calculo do algoritmo do *gradient descent* (ou "descida do gradiente).

#### Gradient Descent

Gradient Descent (descida do gradiente) é um algoritmo utilizado para ajustar os parâmetros de um modelo (como os pesos e vieses de uma rede neural) de forma a minimizar a função de perda, que mede o erro entre as previsões do modelo e os valores reais. Esse algoritmo é essencial para o treinamento de redes neurais e outros modelos de aprendizado de máquina, pois permite que eles “aprendam” e melhorem suas previsões.

#### Otimizadores

Otimizadores são algoritmos que ajustam os parâmetros de um modelo de aprendizado de máquina para minimizar a função de perda (loss function). Eles são essenciais no treinamento de redes neurais, pois definem a forma como o modelo usa o gradiente para atualizar seus pesos e viéses e, assim, melhorar seu desempenho.

Durante o treinamento, o objetivo é reduzir a função de perda. Para isso, o algoritmo calcula o gradiente da função de perda em relação a cada parâmetro (usando o método de backpropagation nas redes neurais). O gradient descent é então aplicado para atualizar esses parâmetros na direção oposta ao gradiente, reduzindo o erro a cada passo.

Os otimizadores definem as regras para esse processo, ajustando o tamanho dos passos de atualização e a forma como o gradiente é usado. Além do gradient descent padrão, existem variações mais avançadas que melhoram o desempenho do modelo, como:

* SGD com Momentum: Acelera o aprendizado acumulando gradientes passados.
* Adam: Ajusta dinamicamente a taxa de aprendizado para cada parâmetro, combinando as vantagens do Momentum e RMSProp.
* RMSProp: Divide o gradiente pelo valor médio das magnitudes de gradiente recentes, ajudando em problemas com grandes oscilações.

Esses otimizadores aprimoram a eficiência do gradient descent, ajudando o modelo a convergir mais rapidamente e a evitar mínimos locais na função de perda. A figura abaixo mostra como funciona esses otimizadores no loop de treinamento do modelo (Stevens, Antiga e Viehmann, 2020, p. 128)

#### Overfitting

Overfitting é um problema comum em deep learning e ocorre quando um modelo aprende excessivamente os detalhes e ruídos do conjunto de dados de treinamento, em vez de captar padrões gerais. Como resultado, o modelo tem um bom desempenho nos dados de treinamento, mas apresenta baixa performance quando é testado em novos dados, pois suas previsões se baseiam em padrões específicos ao conjunto de treino que podem não se generalizar.

Redes neurais profundas possuem uma enorme capacidade de aprendizado devido ao grande número de camadas e parâmetros ajustáveis. Esse poder de representação permite que o modelo aprenda até os menores detalhes dos dados de treinamento, mas também faz com que ele seja mais propenso ao overfitting, especialmente quando:

1. O conjunto de dados de treinamento é pequeno ou não suficientemente variado.
2. O modelo é muito complexo para a tarefa, com muitas camadas e parâmetros em relação à quantidade de dados.

Em deep learning, o overfitting é frequentemente observado quando:

* O erro no conjunto de treinamento continua a diminuir, mas o erro no conjunto de validação começa a aumentar.
* O modelo apresenta alta acurácia no treino, mas baixa acurácia na validação ou em dados de teste.

### Hiperparâmetros

Com base nesses conceitos, podemos definir os principais hiperparâmetros que são importantes para o treinamento de um modelo de aprendizado profundo:

#### Taxa de Aprendizado
A Taxa de Aprendizado (*Learning Rate*) - Grosso modo, é um número que determina o tamanho dos passos que o modelo dá ao ajustar seus parâmetros durante o treinamento. Em outras palavras, ele controla o quanto o modelo "aprende" a cada atualização. Se o learning rate for muito alto, o modelo dá passos grandes, o que pode fazer com que ele "pule" o melhor ponto e nunca chegue a uma solução boa. Se for muito baixo, o modelo dá passos pequenos, o que pode fazer o treinamento ser muito lento e levar mais tempo para chegar a uma solução boa. Géron (final do capítulo 10) diz que um caminho para achar a melhor taxa de aprendizado é "treinar o modelo por algumas iterações, começando com um valor muito baixo (e.g. 10^-5) e ir aumentando gradualmente até um valor muito alto (e.g., 10)". No entanto, nem sempre há tempo para tantas iterações, e alguns modelos já vem com indicações de learning rates (como há para o BERT). Convém testar a maior quantidade de learning rates possíveis e ir verificando como o desempenho do modelo se altera;


#### Iterações

O número de epochs, ou iterações, é um hiperparâmetro. No entanto, não é preciso ajustá-lo se você definir um early-stopping. O early-stopping para o treinamento quando uma certa condição é atingida. Durante o treinamento, o modelo é monitorado em cada época (passagem completa pelos dados), e, se o erro de validação começar a aumentar enquanto o erro de treinamento continua a diminuir, isso indica que o modelo está "decorando" detalhes específicos dos dados de treinamento, em vez de aprender padrões gerais. Early stopping evita isso parando o treinamento no ponto ideal, quando o modelo ainda está bem ajustado para generalizar em dados novos.


#### Otimizador

Além de trocar o learning rate (geralmente um hiperparâmetro dentro do otimizador), você pode trocar o próprio otimizador. Geralmente o ADAM performa bem para a maior parte das tarefas, mas pode ser interessante testar novos valores ([lista de otimizadores](https://pytorch.org/docs/stable/optim.html#algorithms));

#### Batch Size

O batch size é o número de exemplos de treinamento que o modelo processa antes de atualizar seus parâmetros. Em vez de ajustar os pesos após cada exemplo, o modelo agrupa vários exemplos e calcula o erro médio. Depois de processar esse grupo (ou "mini-batch"), ele ajusta os parâmetros de uma só vez.

Um batch size pequeno (por exemplo, 1) pode resultar em atualizações mais rápidas e com mais variação, enquanto um batch size maior (como o conjunto de dados inteiro) tende a gerar atualizações mais estáveis, porém mais lentas. O número padrão costuma ser 32.

O número de epochs, ou iterações, é um hiperparâmetro. No entanto, não é preciso ajustá-lo se você definir um early-stopping. O early-stopping para o treinamento quando uma certa condição é atingida. Durante o treinamento, o modelo é monitorado em cada época (passagem completa pelos dados), e, se o erro de validação começar a aumentar enquanto o erro de treinamento continua a diminuir, isso indica que o modelo está "decorando" detalhes específicos dos dados de treinamento, em vez de aprender padrões gerais. Early stopping evita isso parando o treinamento no ponto ideal, quando o modelo ainda está bem ajustado para generalizar em dados novos.

### Redes Neurais Recorrentes (RNN)

Redes neurais recorrentes (RNNs, do inglês "Recurrent Neural Networks") são um tipo de rede neural projetada para lidar com dados sequenciais, ou seja, dados onde a ordem das informações é importante, como textos, séries temporais e áudio.

Ao contrário das redes neurais tradicionais, onde cada entrada é processada de forma independente, as RNNs possuem conexões recorrentes, ou seja, elas têm conexões que permitem que informações de etapas anteriores da sequência sejam "lembradas" e influenciem a saída atual. Isso é feito através de uma célula de memória, que mantém uma espécie de "estado interno" que é atualizado a cada novo dado da sequência. Isso permite que as RNNs capturem dependências temporais, como padrões em dados de séries temporais ou palavras em uma frase.

No entanto, RNNs tradicionais podem ter dificuldades com sequências muito longas, pois podem esquecer informações importantes devido ao desvanecimento do gradiente. Para superar isso, surgiram variantes como LSTM (Long Short-Term Memory) e GRU (Gated Recurrent Units), que são tipos de RNNs aprimoradas capazes de manter informações por períodos mais longos.

Essas redes são muito usadas em tarefas como:


- **Uso:**
  - Processamento de linguagem natural (NLP).
  - Tradução automática.
  - Previsão de séries temporais.
  - Reconhecimento de voz.

- **Vantagens:**
  - Lida bem com dados sequenciais.
  - Mantém uma memória interna, útil para contextos temporais.
  - Pode ser aplicada a uma variedade de tarefas.

- **Desvantagens:**
  - Dificuldade em lidar com dependências de longo prazo.
  - Propensas a problemas de gradiente que podem dificultar o treinamento.
  - Pode ser computacionalmente intensiva.

## PyTorch

PyTorch é uma biblioteca de aprendizado de máquina de código aberto baseada em Python, desenvolvida pelo Facebook (Meta). Ela é conhecida por sua flexibilidade e é amplamente utilizada em pesquisas e desenvolvimento de modelos de deep learning.

Principais Características:

1. Tensores Dinâmicos:

A principal diferença entre PyTorch e outras bibliotecas é a adoção de tensores dinâmicos. Isso significa que as operações nos tensores podem ser construídas dinamicamente durante o tempo de execução, facilitando a criação de modelos mais dinâmicos e adaptáveis.

2. Diferenciação Automática:

O módulo autograd oferece diferenciação automática. Isso significa que, ao calcular a saída de um modelo, o PyTorch pode automaticamente calcular os gradientes e, assim, facilitar o treinamento de modelos.

3. Construção de Modelos Flexível:

PyTorch permite construir modelos de aprendizado de máquina de maneira intuitiva usando a classe nn.Module. Essa abordagem modular facilita a criação e o entendimento de arquiteturas de modelos complexos.

4. Otimização e Otimizadores:

O módulo torch.optim fornece otimizadores comuns para ajustar os parâmetros dos modelos durante o treinamento. Isso inclui otimizadores como SGD, Adam e outros.

5. Suporte à GPU:

PyTorch oferece suporte nativo à computação em GPU, acelerando significativamente o treinamento de modelos, especialmente em tarefas intensivas computacionalmente.

6. Comunidade Ativa e Documentação Rica:

PyTorch tem uma comunidade ativa e uma documentação rica. A comunidade ativa fornece suporte e recursos adicionais, enquanto a documentação facilita a aprendizagem e o desenvolvimento.

4. Machine/Deep Learning:

* PyTorch:
PyTorch é amplamente usado na implementação de redes neurais, e os tensores são a estrutura fundamental para representar dados e parâmetros de modelos.

````python
import torch

# Criando um tensor
x = torch.tensor([[1, 2, 3], [4, 5, 6]])

# Operações com tensores
y = x + 2
z = torch.matmul(x, x.T)

# Exibindo os resultados
print("Tensor Original:")
print(x)

print("\nOperação de Adição:")
print(y)

print("\nOperação de Produto Matricial:")
print(z)
````

Este exemplo cria um tensor em PyTorch, realiza operações simples (adição e produto matricial) e exibe os resultados. Tenha em mente que tensores em PyTorch compartilham semelhanças conceituais e sintáticas com arrays em Numpy, mas têm vantagens específicas para tarefas de aprendizado de máquina e deep learning.

#### Indexação de Tensores


A indexação de tensores em PyTorch segue padrões semelhantes à indexação de arrays em Python, com algumas particularidades adicionais para suportar operações específicas de aprendizado de máquina e manipulação de dados multidimensionais. Vamos explorar as principais formas de indexação de tensores em PyTorch:

1. Indexação Simples

Você pode acessar elementos individuais de um tensor usando índices. Lembre-se de que a indexação em PyTorch é baseada em zero (o primeiro elemento tem índice 0).

````python
# Criando um tensor
x = torch.tensor([[1, 2, 3], [4, 5, 6]])

# Acessando um elemento específico
elemento = x[1, 2]  # Obtendo o valor 6

print(elemento)
````

2. Slicing (Fatiamento)

Você pode usar a indexação por fatias para extrair partes específicas de um tensor.

````python
# Extraindo uma submatriz
submatriz = x[:, 1:]  # Todas as linhas, da segunda coluna em diante

print(submatriz)
````

3. Indexação Booleana

Usando tensores booleanos para filtrar elementos com base em uma condição.

````python
# Criando um tensor booleano com base em uma condição
condicao = x > 3

# Usando a condição para indexação booleana
elementos_filtrados = x[condicao]


print(elementos_filtrados)
````

#### Juntando tensores

Unir (ou concatenar) tensores em PyTorch é uma operação comum ao trabalhar com dados em aprendizado de máquina e deep learning. PyTorch oferece a função torch.cat() para concatenar tensores ao longo de uma dimensão específica. Vamos explorar como essa operação funciona:

1. Concatenação ao longo de uma dimensão:

A função torch.cat() é usada para concatenar tensores ao longo de uma dimensão específica.

````python
# Criando dois tensores
tensor1 = torch.tensor([[1, 2], [3, 4]])
tensor2 = torch.tensor([[5, 6]])

# Concatenando ao longo da dimensão 0 (linhas)
resultado = torch.cat((tensor1, tensor2), dim=0)

print(resultado)
````

2. Concatenação ao longo de outras dimensões:

Você pode alterar a dimensão ao longo da qual deseja concatenar, especificando o argumento dim.

````python
# Concatenando ao longo da dimensão 1 (colunas)
resultado = torch.cat((tensor1, tensor2.T), dim=1)

print(resultado)
````

3. Concatenação de Tensores com Dimensões Diferentes:

Ao concatenar tensores com diferentes dimensões, é importante que todas as dimensões, exceto a dimensão de concatenação, sejam iguais.

````python
tensor3 = torch.tensor([[7], [8]])

# Concatenando ao longo da dimensão 1 (colunas)
result = torch.cat((tensor1, tensor3), dim=1)
````

Aqui, a operação de concatenação ao longo da dimensão 1 (colunas) é possível porque a dimensão 0 (número de linhas) é a mesma.

````python
print(result)
````

4. Função torch.stack():

Além da função torch.cat(), a função torch.stack() pode ser usada para empilhar tensores ao longo de uma nova dimensão. No entanto, ambos os tensores tem de ter o mesmo tamanho

````python
# Empilhando tensores ao longo de uma nova dimensão
stacked_result = torch.stack((tensor1, tensor2), dim=0)

print(stacked_result)
````

````python
tensor4 = torch.tensor([[4, 3], [2, 1]])

# Empilhando tensores ao longo de uma nova dimensão
stacked_result = torch.stack((tensor1, tensor4), dim=0)

print(stacked_result)
````

Essas operações de concatenação são essenciais ao construir e manipular dados em PyTorch, especialmente ao lidar com lotes de dados ou combinar tensores para entrada em modelos de aprendizado de máquina. Certifique-se de escolher a dimensão correta para a concatenação, dependendo da estrutura específica dos seus dados. ### Datasets e Dataloaders

Em PyTorch, os conceitos de Dataset e DataLoader são fundamentais para trabalhar eficientemente com conjuntos de dados, especialmente ao treinar modelos de aprendizado de máquina. Vamos explorar cada um desses conceitos:

Dataset em PyTorch:
Um Dataset em PyTorch é uma classe que representa um conjunto de dados. Ele fornece uma interface uniforme para acessar as amostras de dados, permitindo que você crie seus próprios conjuntos de dados ou use conjuntos de dados padrão fornecidos por PyTorch.

A classe base para um Dataset em PyTorch é torch.utils.data.Dataset, e ela exige que você implemente os métodos __len__ (retorna o tamanho do conjunto de dados) e __getitem__ (retorna uma amostra específica). [Leia Mais](https://pytorch.org/tutorials/beginner/basics/data_tutorial.html)

````python
from torch.utils.data import Dataset

class CustomDataset(Dataset):
    def __init__(self, data, labels):
        self.data = data
        self.labels = labels

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        sample = {'data': self.data[idx], 'label': self.labels[idx]}
        return sample
````

DataLoader em PyTorch:

Um DataLoader é um iterador que facilita o acesso a conjuntos de dados durante o treinamento de modelos. Ele automaticamente gera mini lotes (batches) de dados e lida com detalhes como embaralhamento e paralelismo de leitura de dados.

O DataLoader é construído sobre um Dataset e simplifica o processo de criação de lotes de dados para o treinamento de modelos. O seguinte exemplo vai dar erro pois ainda não carregamos nenhum banco de dados



```

                        batch_size=64,
                        shuffle=True)
```



Neste exemplo, batch_size define o tamanho de cada mini lote, e shuffle=True indica que os dados serão embaralhados antes de cada época de treinamento.

**Batch Size**

O batch_size é um parâmetro crucial ao treinar modelos de aprendizado de máquina, e sua função está relacionada ao conceito de treinamento em partes, ou lotes (batch training). Vamos entender a importância do batch_size:

* Treinamento em Mini Lotes:

Treinar modelos usando todo o conjunto de dados de uma vez pode ser ineficiente e impraticável, especialmente para grandes conjuntos de dados. Em vez disso, o treinamento é frequentemente realizado em mini lotes, onde o modelo é atualizado com uma parte do conjunto de dados em cada iteração.
Eficiência Computacional:

Usar mini lotes permite aproveitar a eficiência computacional de hardware, como GPUs. Muitas operações em hardware otimizado para aprendizado de máquina podem ser paralelizadas, tornando o treinamento mais rápido ao processar múltiplos exemplos ao mesmo tempo.

* Regularização Estocástica:

O treinamento em mini lotes introduz uma forma de regularização chamada "regularização estocástica". Cada mini lote é uma amostra estocástica do conjunto de dados, o que adiciona uma fonte de variação ao treinamento. Isso pode ajudar a prevenir o overfitting, pois o modelo não vê o conjunto de dados completo em cada iteração.

* Memória Disponível:

Limitações de memória podem ser um fator ao treinar modelos, especialmente em ambientes com recursos limitados. Usar mini lotes permite treinar modelos mesmo quando o conjunto de dados não cabe totalmente na memória.

* Ajuste de Hiperparâmetros:

O tamanho do mini lote é um hiperparâmetro que pode afetar o desempenho do modelo. Em geral, tamanhos de mini lote menores podem levar a atualizações mais frequentes do modelo, enquanto tamanhos maiores podem oferecer uma estimativa de gradiente mais estável.

* Batch Size e Número de Iterações:

O número total de iterações necessárias para uma época de treinamento é determinado pelo tamanho do conjunto de dados dividido pelo tamanho do mini lote. Portanto, o batch_size também influencia a quantidade de processamento necessária para treinar um modelo.

Em resumo, o batch_size é um parâmetro essencial que afeta a eficiência de treinamento, a generalização do modelo e a utilização eficaz de recursos computacionais. Escolher um valor apropriado para o batch_size é uma consideração importante ao treinar modelos de aprendizado de máquina.

````python
import re

# Criando uma função de pré-processamento

def preprocess_text(texto):
    # Use uma expressão regular para encontrar e remover URLs
    texto = re.sub(r'https://\S+', '', texto)

    # Remover quebra de linha
    texto = texto.replace('\\n', ' ')

    # Remover Hashtags
    texto = re.sub(r'#(\w+)', r'\1', texto)

    # Deixar todas as letras minúsculas
    texto = texto.lower()

    # Selecionar apenas letras
    texto = re.findall(r'[a-zéóáêâãõç]+', texto)

    # Juntar o texto, já que o .findall separa em tokens
    texto = " ".join(texto)

    return texto

# Criando a nova coluna transformada
df_ex["texto_processado"] = df_ex['content'].apply(preprocess_text)

# Convertendo a coluna 'Corrected_relevante' em binário
df_ex['Corrected_relevante'] = df_ex['Corrected_relevante'].map({'relevant': 1, 'irrelevant': 0})
````

#### MLP

Vamos seguir o exemplo do ISLP e rodar um modelo com bag-of-words. Primeiro, vamos vetorizar o banco de treino e teste de forma separada como já fizemos anteriormente.

````python
from sklearn.feature_extraction.text import CountVectorizer

# Inicializa o CountVectorizer para Bag-of-Words
vectorizer = CountVectorizer()

# Aplica o BoW nos dados de treino e teste
X_treino_bow = vectorizer.fit_transform(X_treino)
X_teste_bow = vectorizer.transform(X_teste)
````

Vamos criar um modelo de duas camadas de neurônios, como faz o ISLP na página 456.

````python
import torch
import torch.nn as nn # Estruturas
import torch.optim as optim # Otimizadores

class SimpleNNModel(nn.Module):
    def __init__(self, input_dim):
        super(SimpleNNModel, self).__init__()
        # Primeira camada densa
        self.fc1 = nn.Linear(input_dim, 64) # 64 neurônios
        # Segunda camada densa
        self.fc2 = nn.Linear(64, 32) # 32 neurônios
        # Camada de saída
        self.output = nn.Linear(32, 1) # Como é binário, 1 neurônio para a última cada (tabela 10-2 Gerón)
        # Função de ativação
        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()  # Para saída binária; use softmax para múltiplas classes

    def forward(self, x):
        x = self.relu(self.fc1(x)) # Camada densa 1 com ativação ReLU
        x = self.relu(self.fc2(x))  # Camada densa 2 com ativação ReLU
        x = self.sigmoid(self.output(x)) # Output final calculado com a função de ativação sigmoid
        return x
````

Agora, vamos transformar os dados em tensores, e definir dois hiperparâmetros: o BATCH_SIZE e o critério para o cálculo de perda (BCE)

````python
import torch
from torch.utils.data import DataLoader, TensorDataset

# Convertendo dados do Scikit-Learn para tensores PyTorch
X_treino_tensor = torch.tensor(X_treino_bow.toarray(), dtype=torch.float32)
y_treino_tensor = torch.tensor(y_treino.values, dtype=torch.float32)
X_teste_tensor = torch.tensor(X_teste_bow.toarray(), dtype=torch.float32)
y_teste_tensor = torch.tensor(y_teste.values, dtype=torch.float32)

# Criando DataLoaders
BATCH_SIZE = 32
train_data = TensorDataset(X_treino_tensor, y_treino_tensor)
test_data = TensorDataset(X_teste_tensor, y_teste_tensor)
train_loader = DataLoader(train_data, batch_size=BATCH_SIZE, shuffle=True)
test_loader = DataLoader(test_data, batch_size=BATCH_SIZE)



````

Criamos agora a função do loop de treinamento, assim fica mais fácil de alterar os hiperparâmetros. Observe os argumentos colocados na função com atenção, e observe bem onde eles estão dentro da função

````python
import torch

# Pegando as dimensões dos dados
input_dim = X_treino_bow.shape[1]
# Definind o critério de cálculo da Loss
CRITERION = nn.BCELoss()  # Para classificação binária


def train_model(input_dim, criterion, train_loader, epochs=10, device='cpu', model_save_path='model.pth', lr = 0.001):
    # Recriar o modelo e o otimizador a cada execução
    model = SimpleNNModel(input_dim)
    optimizer = optim.Adam(model.parameters(), lr=lr)  # Cria o otimizador para o modelo e define o learning rate

    model.train()  # Coloca o modelo em modo de treino
    for epoch in range(epochs):
        total_loss = 0
        for X_batch, y_batch in train_loader:
            # Enviar os dados para o dispositivo correto
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)

            # Zerar os gradientes
            optimizer.zero_grad()

            # Forward pass
            outputs = model(X_batch).squeeze() # Passando os inputs para o modelo calcular os outputs
            y_batch = y_batch.squeeze()  # Removendo dimensões extras de y_batch se necessário
            loss = criterion(outputs, y_batch) # Calculando a perda com base no critério BCE

            # Backward pass e atualização de pesos
            loss.backward()
            optimizer.step()

            total_loss += loss.item()

        print(f'Epoch [{epoch+1}/{epochs}], Loss: {total_loss/len(train_loader):.4f}')

    # Salvar o modelo após o treinamento
    torch.save(model.state_dict(), model_save_path)
    print(f'Modelo salvo em {model_save_path}')



````

Agora criamos também a função para avaliar o modelo

````python
def evaluate_model(model, criterion, test_loader, model_save_path='model.pth'):
    model.load_state_dict(torch.load(model_save_path, weights_only = True))
    model.eval()
    total_loss = 0
    correct = 0
    total = 0
    with torch.no_grad():
        for X_batch, y_batch in test_loader:
            outputs = model(X_batch).squeeze()
            y_batch = y_batch.squeeze()  # Removendo dimensões extras de y_batch se necessário
            loss = criterion(outputs, y_batch)
            total_loss += loss.item()

            # Predições binárias
            predicted = (outputs > 0.5).float()
            correct += (predicted == y_batch).sum().item()
            total += y_batch.size(0)

    accuracy = correct / total
    print(f'Test Loss: {total_loss/len(test_loader):.4f}, Accuracy: {accuracy:.4f}')
````

Agora vamos rodar o modelo em 10 epochs, veja que estou usando os parâmetros padrões que defini na função

````python
# Rodando o treinamento
train_model(input_dim, CRITERION, train_loader, epochs=10)
````

````python
# Avaliar no conjunto de teste
evaluate_model(model, CRITERION, test_loader)
````

O modelo atinge 91% de acurácia nesse treinamento. Vamos aumentar um pouco o número de epochs e ver como o modelo performa

Uma forma de ver como a mudança de parâmetros altera os resultados na loss é por meio do tensorboard. [Leia Mais](https://pytorch.org/docs/stable/tensorboard.html). Vou adaptar a função de treinamento do modelo para poder ver como a alteração do learning rate afeta nosso modelo simples

````python
from torch.utils.tensorboard import SummaryWriter
import torch
import torch.optim as optim

def train_model_with_lr(input_dim, criterion, train_loader, epochs=10, device='cpu', model_save_path='model.pth', lrs=[0.001, 0.01, 0.1]):
    # Criar o SummaryWriter para o TensorBoard
    writer = SummaryWriter(log_dir='Learning_Rate/runs')  # Salva os logs no diretório 'runs' com o nome "Learning_Rate"

    for lr in lrs:
        # Recriar o modelo e o otimizador a cada execução para cada learning rate
        model = SimpleNNModel(input_dim)
        optimizer = optim.Adam(model.parameters(), lr=lr)  # Define o learning rate

        model.train()  # Coloca o modelo em modo de treino
        print('*'*30)
        print(f'Treinando com Learning Rate = {lr}...')
        print('*'*30)

        for epoch in range(epochs):
            total_loss = 0

            for X_batch, y_batch in train_loader:
                # Enviar os dados para o dispositivo correto
                X_batch, y_batch = X_batch.to(device), y_batch.to(device)

                # Zerar os gradientes
                optimizer.zero_grad()

                # Forward pass
                outputs = model(X_batch).squeeze()  # Passando os inputs para o modelo calcular os outputs
                y_batch = y_batch.squeeze()  # Removendo dimensões extras de y_batch se necessário
                loss = criterion(outputs, y_batch)  # Calculando a perda com base no critério BCE

                # Backward pass e atualização de pesos
                loss.backward()
                optimizer.step()

                total_loss += loss.item()

            # Registrar a perda média para cada época no TensorBoard, para o learning rate atual
            writer.add_scalar(f'Loss/learning_rate_{lr}', total_loss / len(train_loader), epoch)

            print(f'Epoch [{epoch+1}/{epochs}], Loss: {total_loss/len(train_loader):.4f}')

    # Fechar o writer do TensorBoard
    writer.close()

````

````python
lrs = [0.0001,0.001, 0.01, 0.1]  # Lista de valores de learning rate
train_model_with_lr(input_dim=input_dim, criterion=CRITERION, train_loader=train_loader, epochs=100, lrs=lrs)
````
